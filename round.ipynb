### Two numbers X and Y are chosen between 1 and 0. What is the probability that X/Y will rounded to a even number.

Choosing two numbers between 0 and 1 is equavalent to choosing a point from [0,1] x [0,1] rectangle.


```python
import matplotlib.pyplot as plt
import random
```

```python
n = 10000
xPlot = [0]*n
yPlot = [0]*n
```

```python
for i in range(n):
    x = random.uniform(0,1)
    y = random.uniform(0,1)
    rounded = round(y/x)
    if rounded%2 == 0: 
        xPlot[i] = x
        yPlot[i] = y
plt.plot(xPlot, yPlot, 'o', markersize=2)
plt.show()
```

The above plot suggests that the accepted points follow a pattern.
